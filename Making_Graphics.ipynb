```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
prepared_data=pd.read_csv('prepared_data.csv').drop('Unnamed: 0',axis=1)
data = pd.read_csv('Uncleaned_data1.csv').drop('Unnamed: 0',axis=1)
data_significant = pd.read_csv('data_significant.csv').drop('Unnamed: 0',axis=1)
```

```python
f = plt.figure(figsize=(15,5))
ax1 = f.add_subplot(121)
ax2 = f.add_subplot(122)

ax1.set_title("Age Histogram of those who voted for The Republican")
ax1.hist(data[data['voted_for_a']==1]['age'],bins=20)
ax1.set_ylabel("Frequency")
ax1.set_xlabel("Age")
ax1.axvline(x=data[data['voted_for_a']==1]['age'].mean(),color='red')
ax1.text(42,25,r'$\mu:{}$'.format(round(data[data['voted_for_a']==1]['age'].mean(),1)),color='White',
                                            fontsize=15)
ax2 = f.add_subplot(122)
ax2.set_ylabel("Frequency")
ax2.set_title("Age Histogram of those who voted AGAINST The Republican")
ax2.set_xlabel("Age")
ax2.text(60,25,r'$\mu:{}$'.format(round(data[data['voted_for_a']==0]['age'].mean(),1)),color='White',
                                            fontsize=15)
ax2.axvline(x=data[data['voted_for_a']==0]['age'].mean(),color='red')
ax2.hist(data[data['voted_for_a']==0]['age'],bins=20)
plt.savefig('age_hist.png')
plt.show()
```

```python
f = plt.figure(figsize=(12,18))

plt.subplot(321)
plt.title('Voted for the republican \n\n Education level' )
education_for = data[data['voted_for_a']==1]['education'].value_counts()
education_for.name=''
education_for.plot(kind='pie',autopct='%1.1f%%',startangle=0)
plt.axis('tight')

plt.subplot(322)
plt.title('Voted against the republican \n\n Education level')
education_against = data[data['voted_for_a']==0]['education'].value_counts()
education_against.name=''
education_against.plot(kind='pie',autopct='%1.1f%%',startangle=0)
plt.axis('tight')


plt.subplot(323)
plt.title('ethnicity of supporters \n')
ethnic_for = data[data['voted_for_a']==1]['ethnic'].value_counts()
ethnic_for.name=''
ethnic_for.plot(kind='pie',autopct='%1.1f%%',startangle=0)
plt.axis('tight')

plt.subplot(324)
plt.title('ethnicity of opposers \n')
ethnic_against = data[data['voted_for_a']==0]['ethnic'].value_counts()
ethnic_against.name=''
ethnic_against.plot(kind='pie',autopct='%1.1f%%',startangle=0)


plt.subplot(325)
plt.title('Gender of supporters')
sex_for = data[data['voted_for_a']==1]['sex'].value_counts()
sex_for.name=''
sex_for.plot(kind='pie',autopct='%1.1f%%',startangle=90)

plt.subplot(326)
plt.title('Gender of opposers')
sex_against = data[data['voted_for_a']==0]['sex'].value_counts()
sex_against.name=''
sex_against.plot(kind='pie',autopct='%1.1f%%',startangle=90)

#plt.axis('tight')
plt.subplots_adjust(wspace=0.5,hspace=0.3)
plt.savefig('Education_ethnicity_sex.png')
plt.show()
```
